```python
#importing pandas and matplotlib, client/referral csv files and filtering for certain columns within client
#not looking at provinces with less than 20 datapoints

import pandas as pd
import matplotlib.pyplot as plt
client = pd.read_csv("csv files/Client.csv")
referral = pd.read_csv("csv files/Referral.csv")
client_refined = client[["City", "Province", "Origin_Country", "Employment_Status"]]
```

```python
#finding the top three origin countries for number of datapoints and creating plot for % of people employed by country of origin

employed_by_origin_country = client_refined[client_refined["Employment_Status"] == "Employed"].groupby("Origin_Country").count().sort_values("Origin_Country",ascending = True)

num_by_origin_country = client_refined.groupby("Origin_Country").count().sort_values("Origin_Country",ascending = True)

by_percentage_employed = employed_by_origin_country.div(num_by_origin_country).multiply(100).round(2).sort_values(by="Employment_Status",ascending = True)

#by_percentage_employed["Employment_Status"].plot.bar(ylabel = "Proportion of People Satisfactorily Employed", legend=False,title = "Proportion of People Satisfactorily employed by Origin Country",figsize = (12,12))
#plt.savefig("Support Documents/by_percentage_employed_by_origin_country")

#converting column of clients IDs to a dictionary

client_as_dict = client.set_index('Client_ID').T.to_dict('list')

#going through and adding a new column with client into referral dataframe

list_of_origin_countries = []

for i in referral["Client_ID"].tolist():
    list_of_origin_countries.append(client_as_dict[i][5])
    
#referral.insert(2,"Origin_Country",list_of_origin_countries)

#going through and creating a new dataframe by referral status, origin country,interviewed flag and hired flag

refined_referral = referral[["Origin_Country","Referred_Flag","Interviewed_Flag","Hired_Flag"]]

#creating a new dataframe showing the refined_referral dataframe numerically and then creating plots for referral to interview rate and the interview to hired rate by country of origin

origin_country_refined = refined_referral["Origin_Country"].tolist()

referral_flag_refined = []
for i in refined_referral["Referred_Flag"].tolist():
    if i == True:
        referral_flag_refined.append(1)
    elif i == False:
        referral_flag_refined.append(0)

interviewed_flag_refined = []
for i in refined_referral["Interviewed_Flag"].tolist():
    if i == True:
        interviewed_flag_refined.append(1)
    elif i == False:
        interviewed_flag_refined.append(0)

hired_flag_refined = []
for i in refined_referral["Hired_Flag"].tolist():
    if i == True:
        hired_flag_refined.append(1)
    elif i == False:
        hired_flag_refined.append(0)

numerical_referral = pd.DataFrame(list(zip(origin_country_refined,referral_flag_refined,interviewed_flag_refined,hired_flag_refined)),columns = ["origin_country","referral_flag","interviewed_flag","hired_flag"])

referral_to_interviewed_rate = pd.DataFrame((numerical_referral.groupby("origin_country").sum()["interviewed_flag"]/numerical_referral.groupby("origin_country").sum()["referral_flag"]).multiply(100).round(2).sort_values(ascending = True),columns = ["referral_to_interview_conversion"])
#referral_to_interviewed_rate_plot = referral_to_interviewed_rate.plot.bar(ylabel = "Referral to Interview Conversion Rate", legend=False,title = "Referral to Interview Conversion Rate by Origin Country",figsize = (12,12))
#plt.savefig("Support Documents/referral_to_interviewed_rate")

interviewed_to_hired_rate = pd.DataFrame((numerical_referral.groupby("origin_country").sum()["hired_flag"]/numerical_referral.groupby("origin_country").sum()["interviewed_flag"]).multiply(100).round(2).sort_values(ascending = True),columns = ["interview_to_hired_conversion"])
#interviewed_to_hired_rate_plot = interviewed_to_hired_rate.plot.bar(ylabel = "Interview to Hired Conversion Rate", legend=False,title = "Interview to Hired Conversion Rate by Origin Country",figsize = (12,12))
#plt.savefig("Support Documents/interviewed_to_hired_rate")
```
